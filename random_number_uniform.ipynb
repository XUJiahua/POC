## 验证下随机数是否是均匀分布的

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
sample_counts = [100, 1000, 10000, 100000]
low = 0
high = 1000
```

### 均匀分布

https://docs.scipy.org/doc/numpy/reference/generated/numpy.random.uniform.html


参数：最大值，最小值

![uniform_disturibution](https://docs.scipy.org/doc/numpy/_images/math/30892d6b0dfe8db2ebc0248d3bfe4aba4ac65171.svg)


### 正态分布

https://docs.scipy.org/doc/numpy/reference/generated/numpy.random.normal.html

参数：中心值、标准差

![normal_disturibution](https://docs.scipy.org/doc/numpy/_images/math/192d06ecdfd9d671a4ab4dd942ddf8e20173e111.svg)



```python
plt.figure(figsize=(20,18))

for i, sc in enumerate(sample_counts):
    plt.subplot(len(sample_counts), 2, i*2+1)
    plt.title("sample counts: {}, range from {} ~ {}".format(sc, low, high))
    s = np.random.uniform(low, high, size=sc)
    plt.hist(s, bins=(high-low))
    
    plt.subplot(len(sample_counts), 2, i*2+2)
    # 使用上述数据的均值和标准差构建高斯/正态分布
    mu = s.mean()
    sigma = s.std()
    plt.title("sample counts: {}, mu: {}, sigma: {}".format(sc, mu, sigma))
    s = np.random.normal(mu, sigma, size=sc)
    plt.hist(s, bins=(high-low))
```

### randint以及语言built-in API的随机数生成也是均匀分布

https://docs.scipy.org/doc/numpy/reference/generated/numpy.random.randint.html

```python
plt.figure(figsize=(10,18))

for i, sc in enumerate(sample_counts):
    plt.subplot(len(sample_counts), 1, i+1)
    plt.title("sample counts: {}, range from {} ~ {}".format(sc, low, high))
    
    s = [np.random.randint(low, high) for _i in range(sc)]
    plt.hist(s, bins=(high-low))
    
```
